```python
import numpy as np
from keras.layers import Dense
from keras.models import Sequential
from sklearn.model_selection import train_test_split
```

```python
bits=8
```

```python
def gennos(v):
    x1=[]
    for i in range(254):
        y1=[]
        z=bin(i + v)[2:].zfill(bits)
        for j in z:
            y1.append(int(j))
        x1.append(y1)
    return np.array(x1)
```

```python
def dataset():
    return gennos(0),gennos(1)
```

```python
x2,y2=dataset()
X_train,X_test,Y_train,Y_test = train_test_split(x2,y2,test_size=0.28,random_state=42)
```

```python
print(X_train,Y_train)
print(X_test,Y_test)
print(X_train.shape,Y_train.shape)
```

```python
model=Sequential()
model.add(Dense(units=128,input_dim=bits,activation='sigmoid'))
model.add(Dense(units=64,activation='sigmoid'))
model.add(Dense(units=bits,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(X_train,Y_train,epochs=500,batch_size=5)
```

```python
model.evaluate(X_test,Y_test,verbose=1,batch_size=128)
```

```python
def next(val):
    if val <= 254:
        out_l = []
        inp_l = []
        b = bin(val)[2:].zfill(8)
        for k in b:
            inp_l.append(int(k))
        out_l.append(inp_l)
        out_l = np.array(out_l)
        classes = model.predict(out_l)
        output = []
        for u in classes:
            li = []
            for z in u:
                if z >= 0.1:
                    li.append(1)
                else:
                    li.append(0)
            output.append(li)
        output = np.array(output)
        print("Input:", out_l, "output:", output)
    else:
        print("Exceeds")
```
